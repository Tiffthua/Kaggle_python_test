## loading the data 

```python
df = pd.read_csv(r'orders.csv')
df.head(7)
```

```python
df.isnull().sum()
```

```python
### Categorical Exploration
for col in df.select_dtypes(include = 'object'):
    if col.lower() != 'order date':
        print('unique values for column', col)
        print(df[col].unique())
```

```python
import numpy as np
```

```python
## Replace Invalids
df['Ship Mode'] = df['Ship Mode'].replace(['Not Available','unknown'], np.nan) 
df.isnull().sum()
```

```python
##  Rename Columns
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace (" ", "_")
df.head(30)
```

```python
## Create Derived Columns
## Compute a discount column as list_price * discount_percent / 100. 
df['discount'] = df['list_price'] * df['discount_percent'] / 100

df.head(30)
```

```python
##  Compute a selling_price column as list_price - discount.
df['selling_price'] = df['list_price'] - df['discount']
```

```python
## Compute a profit column as selling_price - cost_price. 
df['profit'] = df['selling_price'] - df['cost_price']
```

```python
## Drop Unneeded Columns
df = df.drop(['cost_price', 'discount_percent', 'list_price'], axis = 1)

df.head(30)
```

```python
## Datetime Conversion
df['order_date'] = pd.to_datetime(df['order_date'])

df.head(30)
```
