```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd
```

## Univariate plotting (Histogram, Box plot, Bar plot)

```python
### loading the dataset
orders = pd.read_csv(r'orders.csv')
orders.head()
```

```python
## histogram plot Distribution of Numeric Data 
sns.histplot(orders['Discount Percent'], kde=True, color='green')
plt.title('Distribution of Discount Percent')
plt.xlabel('Discount Percent')
plt.ylabel('Count')

plt.show()
```

```python
### Box plot Shows Outliers and Spread 
sns.boxplot(orders['Quantity'], color= 'Purple')
plt.title('Boxplot of Quantity')
plt.xlabel('median')

plt.show()
```

```python
### bar plot Categorical Frequency 
sns.countplot(y ='Category', data=orders, palette = 'pastel')
plt.title('Number of Categories')
plt.xlabel('Category')
plt.ylabel('Count')

plt.show()
```

## Bivariate Analysis(Scatter plot, Box plot, Grouped Bar plot )

```python
orders.head()
```

```python
### Scatter plot Numeric vs Numeric
sns.scatterplot(x = 'cost price', y = 'Quantity', data=orders)
plt.title('cost price vs list price')
plt.xlabel('Cost Price')
plt.ylabel('Quantity')

plt.show()
```

```python
### Box plot Categorical vs Numeric 
sns.boxplot(x = 'Category', y = 'Quantity', data=orders, color='purple')
plt.title('Cost price by sub category')
plt.xlabel('Category')
plt.ylabel('Quantity')

plt.show()
```

```python
### Grouped Bar Plot — Categorical vs Categorical 
sns.countplot(x = 'Segment', hue= 'Region', data=orders, palette= 'Set1')
plt.title('Ship mode by Region')
plt.xlabel('Segment')
plt.ylabel('Count')

plt.show()
```

### Multivariate Analysis (Scatter plot, Pair Plot, Heatmap,)

```python
orders.head()
```

```python
###  Hue and Style in Scatter Plot 
sns.scatterplot(x = 'Quantity', y = 'List Price', hue= 'Segment', style= 'Region', data=orders)
plt.title('List price vs Quantity by segment and region')

plt.show()
```

```python
###  Pair Plot — Matrix of Scatter Plots 
sns.pairplot(orders[['cost price', 'Quantity', 'Ship Mode', 'Region']], hue= 'Region')
plt.suptitle('Pairwise relationship', y= 1.02)

plt.show()
```

```python
### Heatmap — Correlation Matrix 
correlation = orders[['cost price', 'Quantity', 'Discount Percent']].corr() 
sns.heatmap(correlation, annot=True, cmap='coolwarm') 
plt.title("Correlation Heatmap Orders") 
plt.show() 
```

## Visualizing Categorical Data 

```python
orders.head()
```

```python
###  Pie Chart
orders['Region'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette("pastel")) 
plt.title("Region pie chart") 
plt.ylabel("") 
plt.show() 
```

```python
### Violin Plot — Distribution + Density 
sns.violinplot(x= 'Region', y= 'Discount Percent' ,data=orders, palette= ['red', 'purple', 'yellow', 'green'])
plt.title('Regional distribution by discount percentage')

plt.show()
```

```python
### Swarm Plot — Exact Data Points 
sns.swarmplot(x='Quantity', y='Region', data=orders, palette='Accent') 
plt.title("Quantity by Segment (Exact Values)") 

plt.show(10) 
```
